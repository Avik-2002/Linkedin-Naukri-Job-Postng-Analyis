```python
job_list=[]

for job_id in id_list:
    job_url = f"https://www.linkedin.com/jobs-guest/jobs/api/jobPosting/{job_id}"
    job_response = requests.get(job_url)
    print(job_response.status_code)
    job_soup = BeautifulSoup(job_response.text, "html.parser")

    job_post={}
    try:
        job_post["job_title"] = job_soup.find("h2", {"class":"top-card-layout__title font-sans text-lg papabear:text-xl font-bold leading-open text-color-text mb-0 topcard__title"}).text.strip()
    except:
        job_post["job_title"] = None
    try:
        job_post["company_name"] = job_soup.find("a", {"class": "topcard__org-name-link topcard__flavor--black-link"}).text.strip()
    except:
        job_post["company_name"] = None
    try:
        job_post["time_posted"] = job_soup.find("span", {"class": "posted-time-ago__text topcard__flavor--metadata"}).text.strip()
    except:
        job_post["time_posted"] = None
    try:
        job_post["num_applicants"] = job_soup.find("span", {"class": "num-applicants__caption topcard__flavor--metadata topcard__flavor--bullet"}).text.strip()
    except:
        job_post["num_applicants"] = None
    
job_list.append(job_post)
    
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
chrome_options = Options()
chrome_options.add_argument("--headless") 

service = Service() 
driver = webdriver.Chrome()
job_list = []
for i in range(2, 10):

        url =f"https://www.naukri.com/python-developer-data-analyst-data-processing-jobs-{i}?k=python+developer%2C+data+analyst%2C+data+processing&experience=0&nignbevent_src=jobsearchDeskGNB"
        driver.get(url)
        time.sleep(5)
        data = driver.page_source
        soup = BeautifulSoup(data, 'html.parser')

        all_jobs = soup.find_all("div",class_="srp-jobtuple-wrapper")
        # print(all_jobs)

        
        for job in all_jobs:
                title_tag = job.find("a", class_="title")
                title = title_tag.get_text(strip=True) if title_tag else None
                
                company_tag = job.find("a", class_="comp-name")
                company = company_tag.get_text(strip=True) if company_tag else None
                
                rating_tag = job.find("a", class_="rating")
                rating = rating_tag.get_text(strip=True) if rating_tag else None
                
                location_tag = job.find("span", class_="ni-job-tuple-icon ni-job-tuple-icon-srp-location loc")
                location = location_tag.get_text(strip=True) if location_tag else None
                
                experience_tag = job.find("span", class_="ni-job-tuple-icon ni-job-tuple-icon-srp-experience exp")
                experience = experience_tag.get_text(strip=True) if experience_tag else None
                
                link_tag = job.find("a", class_="title")
                link = link_tag['href'] if link_tag and link_tag.has_attr('href') else None


                skills_tag = job.find("ul", class_="tags-gt")
                if skills_tag:
                        skills = [skill.get_text(strip=True) for skill in skills_tag.find_all("li",class_="dot-gt tag-li")]
                else:
                        skills = []
                
                job_info = {
                        "job_title": title,
                        "company_name": company,
                        "rating": rating,
                        "location": location,
                        "experience_required": experience,
                        "skills_required": skills,
                        "link": link,
                }
                
                job_list.append(job_info)
print(json.dumps(job_list, indent=4))

df_naukri = pd.DataFrame(job_list)
```
